```python
import pandas as pd
data = pd.read_csv("/content/SoftwareProject_DataSet.csv")
```

```python
data.head(n=10)
```

```python
grouped_by_project_type = data.groupby("Idioma de programación")
```

```python
stratified_sample = grouped_by_project_type.sample(n=5, random_state=42)
print("Muestra estratificada por Idioma de programación")
stratified_sample_sorted = stratified_sample.sort_values("Idioma de programación")
display(stratified_sample_sorted)
```
